```python
# 1: Login to your Instagram Handle
 # 1: Submit with sample username and password
from selenium.webdriver.common.keys import Keys                            #provide keys in the keyboard like RETURN, F1, ALT
from selenium.webdriver.support.ui import Select                           #to select the option html attribute
from selenium.webdriver.support.ui import WebDriverWait                    #To use implcit and explicit wait
from selenium.webdriver.support import expected_conditions as EC           #use in explicitly wait
from selenium.webdriver.common.by import By                                #to select the attribute by Class,link_text
from selenium import webdriver                                             #import web Driver
import time                                                                #it use in wait
from bs4 import BeautifulSoup                                              #work with attribute 
from selenium.common.exceptions import TimeoutException

from selenium import webdriver
driver=webdriver.Chrome('/webdriver/chromedriver')
d=driver.get('https://www.instagram.com/')
wait = WebDriverWait(driver, 10)
time.sleep(4)
u_name=driver.find_element_by_name('username')
u_name.send_keys('USERNAME')
pas=driver.find_element_by_name('password')
pas.send_keys('PASSWORD')
pas.submit()
```

```python
# 2: Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”
 # 1: Note : Make sure to avoid printing hashtags
d1=driver.find_elements_by_xpath('//div[@class="mt3GC"]/button')
d1[1].click()
time.sleep(2)
d2=driver.find_element_by_xpath('//div[contains(@class,"LWmhU")]/input')
d2.send_keys('food')
d2.click()
time.sleep(2)
d=driver.find_element_by_class_name('drKGC')
d.click()
data=BeautifulSoup(driver.page_source)
d1=data.find_all(class_='Ap253')
for i in d1:
    if i.string[0]!='#':
        print(i.string)
```

```python
# 3 Searching and Opening a profile using 
 #Open profile of “So Delhi” 
d2=driver.find_element_by_xpath('//div[contains(@class,"LWmhU")]/input')
d2.send_keys('So Delhi')
d2.click()
time.sleep(3)
dt=driver.find_element_by_class_name('z556c')
dt.click()
```

```python
# 4 follow/Unfollow given handle - 
#1: Open the Instagram Handle of “So Delhi”
#2: Start following it. Print a message if you are already following
#3: After following, unfollow the instagram handle. Print a message if you have already unfollowed.
# For Follow the user
d1=driver.find_elements_by_xpath('//div[@class="mt3GC"]/button')
d1[1].click()
d2=driver.find_element_by_xpath('//div[contains(@class,"LWmhU")]/input')
d2.send_keys('So Delhi')
d2.click()
time.sleep(3)
dt=driver.find_element_by_class_name('z556c')
dt.click()
time.sleep(2)
d3=driver.find_element_by_xpath('//div[contains(@class,"Igw0E")]//span/button')
if d3.text=='Follow':
    d3.click()
    time.sleep(3)
else:
    print('You are already following')
```

```python
# For unfollow
d1=driver.find_elements_by_xpath('//div[@class="mt3GC"]/button')
d1[1].click()
d2=driver.find_element_by_xpath('//div[contains(@class,"LWmhU")]/input')
d2.send_keys('So Delhi')
d2.click()
time.sleep(3)
dt=driver.find_element_by_class_name('z556c')
dt.click()
time.sleep(2)
d3=driver.find_element_by_xpath('//div[contains(@class,"Igw0E")]//span/button')
if d3.text=='Follow':
    d3.click()
    time.sleep(3)
else:
    print('You are already following')
d3.click()
time.sleep(2)
de=driver.find_element_by_xpath('//div[@class="mt3GC"]/button')
if de.text=='Unfollow':
    de.click()
driver.back()
```

```python
# 5 Like/Unlike posts
#1: Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
#2: Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.
# For Liking top 30 Posts
d2=driver.find_element_by_xpath('//div[contains(@class,"LWmhU")]/input')
d2.send_keys('dilsefoodie')
d2.click()
time.sleep(2)
d2=driver.find_element_by_class_name('z556c')
d2.click()


driver.execute_script('window.scrollBy(0,550)')
time.sleep(4)
driver.execute_script('window.scrollBy(0,600)')
time.sleep(5)
d2=driver.find_elements_by_class_name('_9AhH0')
for i in d2[:30]:
    i.click()
    time.sleep(3)
    d3=driver.find_element_by_xpath('//span[@class="fr66n"]/button')
    data=BeautifulSoup(driver.page_source)
    dh=data.find(class_="fr66n")
    res=str(dh.svg['aria-label'])
    if res=='Unlike':
        print('Already liked')
    else:
        d3.click()
    time.sleep(3)
    d4=driver.find_element_by_xpath('//div[contains(@class,"qJPeX")]/button')
    
    d4.click()
    time.sleep(2)
```

```python
# For unliking the top 30 posts

d2=driver.find_element_by_xpath('//div[contains(@class,"LWmhU")]/input')
d2.send_keys('dilsefoodie')
d2.click()
time.sleep(4)
d2=driver.find_element_by_class_name('z556c')
d2.click()

driver.execute_script('window.scrollBy(0,550)')
time.sleep(4)
driver.execute_script('window.scrollBy(0,600)')
time.sleep(5)
d2=driver.find_elements_by_class_name('_9AhH0')
for i in d2[:30]:
    i.click()
    time.sleep(3)
    d3=driver.find_element_by_xpath('//span[@class="fr66n"]/button')
    data=BeautifulSoup(driver.page_source)
    dh=data.find(class_="fr66n")
    res=str(dh.svg['aria-label'])
    if res=='Like':
        print('Already Unliked')
    else:
        d3.click()
    time.sleep(3)
    d4=driver.find_element_by_xpath('//div[contains(@class,"qJPeX")]/button')
    
    d4.click()
    time.sleep(2)
```

```python
# 7 ceck the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
# 1: If You have already seen the story.
# 2: Or The user has no story.
# 3: Or View the story if not yet seen.
d1=driver.find_elements_by_xpath('//div[@class="mt3GC"]/button')
d1[1].click()
count=0
for i in range(3):
    search = driver.find_element_by_xpath('//div[contains(@class,"LWmhU")]/input')      #input box  
    search.send_keys('codingninjas')
    first_search = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'z556c')))  #fisrt profile
    first_search.click()
    print('Not checking the story yet and now i am going to check it')
    time.sleep(2)
    try:
        a = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'h5uC0')))
        if count==1:
            print("Already seen the Story.")
            driver.back()
        else:
            a = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_2dbep')))              #for story check button
            a.click()
            print("Seen the story")
            driver.back()
            count+=1

    except TimeoutException:
        print("User don't has Story.")
        time.sleep(2)
    driver.back()
```
